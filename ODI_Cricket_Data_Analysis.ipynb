##Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

###Count of players by team

```python
batter['team'].nunique()

plt.figure(figsize=(8, 6))
sns.countplot(x='team', data=batter, palette="viridis", hue='team', legend=False)
plt.title('Count of Players by Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

###Total Runs

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='total_runs',data=batter)
plt.title('Total Runs')
plt.show()
```

##Bivariate analysis

###Total runs Vs total_matches_played

```python
sns.scatterplot(data = batter, x='total_matches_played', y='total_runs', hue='team')

plt.xlabel('Total Matches Played')
plt.ylabel('Total Runs')
plt.title('Total Runs Vs Total Matches Played')
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.show()
```

### Total runs Vs Top 10 batsman

```python
top_batsman = batter.sort_values(by= 'total_runs', ascending=False).head(10)

plt.figure(figsize=(8,6))
sns.barplot(x=top_batsman['player_name'], y=top_batsman['total_runs'], data = top_batsman, hue=top_batsman['player_name'], palette="magma")
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 10 Batsman with highest runs')
plt.xticks(rotation=90)
plt.show()
```
